```
#Import the libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
```

```
#Load the data
from google.colab import files
files.upload()
```

```
#Store the data into a variable/data frame
df = pd.read_csv('LVGO.csv')
#Set date as index
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
#Name index
df.index.name = 'Date'
#Show the data
df
```

```
#Manipulate data/create target column
df['Price_Up'] = np.where(df['Close'].shift(-1) > df['Close'], 1,0)

#Remove Date column
df = df.drop(columns=['Date'])
#See Data
df
```

```
#Split Data set into a feature dataset and a target data set
X = df.iloc[:, 0:df.shape[1]-1].values
Y = df.iloc[:,df.shape[1]-1].values
```

```
#Split data agin into 80% training  and 20% testing data sets
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

```
#Create and Train the model (DecisionTreeClassifier)
tree = DecisionTreeClassifier().fit(X_train, Y_train)

#Show how well the model did on the test data
print(tree.score(X_test, Y_test))
```

```
#Show the models predictions
tree_predictions = tree.predict(X_test)
print(tree_predictions)
```
